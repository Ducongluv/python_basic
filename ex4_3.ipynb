## 불변 객체의 공유 참조
- int, float, bool, str, tuple, frowezset 등이 불변 객체

```python
a = 10
b = a
print(a, b)
print(a is b)
print(a == b)
```

```python
s1 = "Python"
# s2 = "Pytho"

# s2 = s2 + "n" # 이 경우 s1 is s2는 False가 나온다 왜냐면 s2는 언제든 가변이 될수있기 때문

s2 = "Pytho" + "n" # True, Pytho + n은 무조건 Python이지만 s2는 언제든 Pytho가 아닌 다른게 될 수 있기 때문에 새로운 공간을 만들었음 그 위에 False가 나오는 이유

print(s1 is s2) 
print(s1 == s2) 
```

```python
nums1 = [10, 20, 30]
nums2 = nums1

print(nums1 is nums2) 
print(nums1 == nums2)

nums1[0] = 999
print(nums1, nums2)

print(nums1 is nums2) 
print(nums1 == nums2)
```

```python
nums1 = [10, 20, 30]
nums2 = [10, 20, 30]

print(nums1 is nums2) # 가변데이터의 경우 공동 참조 되지 않음
print(nums1 == nums2)
```

```python
a = [1, 2, 3]
b = a

print(a is b)
print(a == b)

print(id(a), id(b))
```

```python
a = 2
b = 1 + 1

print(a is b)
print(a == b)

print(id(a), id(b))
```
