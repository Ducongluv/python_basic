```python
# 난수 생성을 제공하는 random 모듈
import random

print(random.random()) # 0~1 사이 실수
print(random.randint(1, 10)) # 1~ 10 정수
print(random.choice([1, 2, 3])) # 랜덤 선택
print(random.choice(range(1, 11)))
```

```python
number = range(1, 46)

balls = list(range(1, 46))
print(balls)

for i in range(0, 1000000):
    temp = balls[0]
    r_num = int(random.random() * len(balls)) # 0 ~ 44
    # math module import 시 r_num = math.floor(random.random() * len(balls))
    balls[0] = balls[r_num]
    balls[r_num] = temp

print(balls)
print(balls[:6])
```

```python
for i in range(10):
    lotto = random.sample(range(1, 46), 6)
    lotto.sort()
    print(f"{i}번째 로또 번호 : {lotto}")
```

```python
def my_lotto_create(a):
    result = []
    for i in range(a):
        lotto = random.sample(range(1, 46), 6)
        lotto.sort()
        #print(f"{i + 1}번째 로또 번호 : {lotto}")
        result.append(lotto)
    return tuple(result)
```

```python
result = my_lotto_create(10)
print(result)
```

```python
def dice_cast():
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    return d1 + d2

# person1_dice = {"name" : "홍길동", "dice_value" : 0}
# person1_dice["dice_value"] = dice_cast()
person1_dice = dice_cast()
person2_dice = dice_cast()

if person1_dice == person2_dice:
    print("비겼다")
elif person1_dice > person2_dice:
    print(f"person1 승리. person1 : {person1_dice}, person2 : {person2_dice}")
elif person1_dice < person2_dice:
    print(f"person2 승리. person1 : {person1_dice}, person2 : {person2_dice}")

```

```python
# 컬렉션 관련 collections 모듈

from collections import Counter
# 수학 관련 기능 중 내장함수에 없는 기능은 math 모듈
# 컬렉션 다루는 기능 중 내장함수에 없는 기능은 collections 모듈
data = ["a", "a", "x", "a", "z", "z", 4]
print(Counter(data)) # Counter({'a': 3, 'z': 2, 'x': 1, 4: 1})
```
