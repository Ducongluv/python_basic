## 결측치 확인


```python
missing_count = 0

for order in orders:
    if order["구매금액"] is None:
        missing_count += 1

print("결측치 개수 : ", missing_count)
```

## 결측치 처리

```python
prices = []

for order in orders:
    if order["구매금액"] is not None:
        prices.append(order["구매금액"])

avg = sum(prices) / len(prices)

for order in orders:
    if order["구매금액"] is None:
        order["구매금액"] = avg

orders
```

## 이상치 탐지

```python
# 정상 범위 0 ~ 1000000
for order in orders:
    price = order["구매금액"]

    if price < 0 or price > 1000000:
        print(order)
```

```python
# 이상치 처리
for order in orders:

    if order["구매금액"] > 1000000:
        order["구매금액"] = 1000000
    elif order["구매금액"] < 0:
        order["구매금액"] = 0

orders
```

## 특성 추출

```python
# 가입연수
from datetime import datetime

current_year = datetime.now().year

for order in orders:
    rrn = order["가입일"]
    join_year = int(rrn[:4])

    order["가입연수"] = current_year - join_year

orders
```

```python
# VIP 여부 생성

for order in orders:
    if order["구매금액"] >= 200000:
        order["회원등급"] = "VIP"
    else:
        order["회원등급"] = "일반"

orders
```

## 지역별 평균 구매금액

```python
# 지역 확인(서울, 부산, 대구, 광주, 대전)
for order in orders:
    print(order["지역"])
```

```python
seoul_sum = 0
seoul_count = 0

busan_sum = 0
busan_count = 0

daegu_sum = 0
daegu_count = 0

gwangju_sum = 0
gwangju_count = 0

daejeon_sum = 0
daejeon_count = 0

for order in orders:

    if order['지역'] == "서울":
        seoul_sum += order["구매금액"]
        seoul_count += 1
    elif order['지역'] == "부산":
        busan_sum += order["구매금액"]
        busan_count += 1
    elif order['지역'] == "대구":
        daegu_sum += order["구매금액"]
        daegu_count += 1
    elif order['지역'] == "광주":
        gwangju_sum += order["구매금액"]
        gwangju_count += 1
    else:
        daejeon_sum += order["구매금액"]
        daejeon_count += 1

print("서울 평균:", seoul_sum / seoul_count)
print("부산 평균:", busan_sum / busan_count)
print("대구 평균:", daegu_sum / daegu_count)
print("광주 평균:", gwangju_sum / gwangju_count)
print("대전 평균:", daejeon_sum / daejeon_count)
```

## 회원등급별 평균 구매금액
