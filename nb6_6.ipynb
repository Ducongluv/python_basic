## 가입연수별 그룹 분석

```python
# 가입연수별 평균 금액
price1, price2, price3 = 0, 0, 0 # 0~2, 3~5, 6년 이상 그룹별 구매금액 누적
count1, count2, count3 = 0, 0, 0 # 0~2, 3~5, 6년 이상 그룹별 카운트 누적

for order in orders:
    year = order["가입연수"]
    
    if year <= 2: # 가입연수 0~2
        price1 = price1 + order["구매금액"]
        count1 = count1 + 1
    
    elif year <= 5: # 가입연수 3~5
        price2 = price2 + order["구매금액"]
        count2 = count2 + 1

    else: # 가입연수 6
        price3 = price3 + order["구매금액"]
        count3 = count3 + 1

print(f"0~2년 : {int(price1 / count1)}")
print(f"3~5년 : {int(price2 / count2)}")
print(f"6년 이상 : {int(price3 / count3)}")
```

## 구매금액 높은 고객 TOP 3

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# 예시 배열
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
print("Sorted array is:", arr)
```

```python
result = []

for order in orders:
    result.append(order)


# bubble sort

n = len(result)
for i in range(n):
    for j in range(0, n - i -1):
        if result[i]["구매금액"] > result[j]["구매금액"]:
            temp = result[i]
            result[i] = result[j]
            result[j] = temp


print(result[0])
print(result[1])
print(result[2])
```
