## 함수 호출시 "불변객체 인수 vs 가변객체 인수" 사용 시 차이 

```python
# 불변 객체를 인수로 전달
# int, float, str, tuple, bool 등
def func(x):
    x += 10
    print("함수 안 : ", x, id(x))

a= 10

func(a)

print("함수 밖 : ", a, id(a))

# 함수 안의 x만 변경, 함수 밖의 a 는 그대로
# id() 값이 다르다
```

```python
# 가변 객체를 인수로 전달
# list, dic, set 등
def funs(nums):
    nums.append(4)
    print("함수 안 : ", nums, id(nums))

data = [1, 2, 3]

func(data)

print("함수 밖:", data, id(data))

# id() 값이 동일
```

```python
def func(nums):
    print("새로운 값 대입 전:", nums, id(nums))
    nums = [100, 200]
    print("새로운 값 대입 후:", nums, id(nums))

data = [1, 2, 3]
func(data)
print(data, id(data))

# 새로운 값 대입 전: [1, 2, 3] 2514266872192
# 새로운 값 대입 후: [100, 200] 2514266954944
# [1, 2, 3] 2514266872192
```

## 함수 코드블록안에서 전역변수의 사용

​

- 함수 코드블록 안에 전역변수를 사용하는 것은 좋지 않은 방법이다.  

- 대부분의 언어들은 함수안에 전역변수를 사용하는 것을 코딩 규칙으로 사용하지 않도록 한다.

- 왜 문법으로 막지 않느냐? "언어가 모든 것을 금지하기보다, 개발자가 책임 있게 사용하도록 한다"는 것이 프로그래밍 언어의 철학이다.

​

※ 전역변수는 문법적으로 사용할 수 있지만, 함수가 전역변수에 의존하면 재사용성과 유지보수성이 떨어진다. 따라서 필요한 값을 매개변수로 전달하는 것이 가장 좋은 방법이며, 전역변수는 주로 상수나 프로그램 전체에서 공유하는 설정값의 읽기용도로만 사용하는 것이 좋다.


```python
# 전역 변수 : 함수 정의와 같은 코드 레벨에 있는 변수
x = 10 

def fun_global():
    print(x) # x 전역변수를 사용

def fun_local():
    x = 333
    print(x) # 지역변수 x가 존재하기 때문에 지역변수 x를 사용


fun_global()
fun_local()
```

## 사용자 정의 함수 실습

```python
def my_intersect(a, b) -> list:
    """
    a : 리스트
    b : 리스트
    반환타입 : 리스트
    내용 : 두 개의 리스트의 중복되는 값만을 새로운 리스트로 반환
    """
    result = []
    
```

```python
# 두 시퀀스의 교집합을 구하는 함수

def my_intersect(a, b):
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result

print(my_intersect([1, 2, 3], [2, 3, 5]))
print(my_intersect((1, 2, 3), (2, 3, 5)))
print(my_intersect({1, 2, 3}, {2, 3, 5}))
print(my_intersect({"id" : "abc", "sn" : "1111"}, {"id" : "xyz", "pw" : "1234"}))
```

```python
# a에는 있고 b에는 없는 요소

def my_difference(a, b):
    result = []
    for e in a:
        if e not in b:
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 합집합

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
# value가 시퀀스 안에 있는지 확인

def my_contains(data, value):

    for e in data:
        if e == value:
            return True

    return False


print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

```python
def add(a : int, b : int) -> int:
    return a + b
print(add("안녕", "길동"))
```

```python
# 특정 값이 몇 번 나오는지 계산

def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1

    return count


print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

```python
# 중복 제거

def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```

## 1급 함수
1급 함수(First-Class Function)​란 함수를 일반 데이터(객체)처럼 다룰 수 있는 함수를 말한다.

파이썬에서는 함수도 객체이므로 1급 함수를 지원한다.

즉, 함수를 다음과 같이 사용할 수 있다.

​

1) 변수에 저장할 수 있다. 

2) 함수의 인수(argument)로 전달할 수 있다. 

3) 함수의 반환값(return)으로 사용할 수 있다. 

4) 자료구조(list, dict 등)에 저장할 수 있다.

```python
# 변수에 저장하기

def hello():
    print("Hello")

f = hello      # 함수 자체를 저장 (괄호 없음)

f()
```

```python
# 함수의 인수로 전달하기

def square(x):
    return x * x

def test(fun):
    return fun(5)

print(test(square))
```

```python
# 함수를 반환하기

def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()

(outer())()
```

```python
# 자료 구조에 저장하기

def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y


def sub(x, y):
    return x - y

calc = {
    "+" : add,
    "-" : sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```
