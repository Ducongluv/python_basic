## LIST

```python
# 연결(+) 연산
a = [1, 2, 3]
b = [4, 5]
result = a + b
print(result) 
print(a)
print(b)
```

```python
# 반복(*) 연산
nums = [10, 20]
result = nums * 3
print(result)
```

```python
# in, not in 연산
fruits = ["apple", "banana", "orange"]
print( "banana" in fruits)

print( "grape" in fruits)
print( "grape" not in fruits)
```

```python
# 인덱싱
nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)
```

```python
# 슬라이싱
nums = [10, 20, 30, 40, 50]

print(nums[1:4])
print(nums[:3])
print(nums[2:])
print(nums)
```

```python
# 2칸씩 건너뛰기
nums = [10, 20, 30, 40, 50, 60, 70 ,80]
print(nums[::2])

print(nums[1::2])
```

```python
# 역순 출력
nums = [10, 20, 30, 40, 50]
print(nums[::-1])
```

```python
# 뒤에서 3개 추출
nums = [10, 20, 30, 40, 50]
print(nums[-3:])
```

```python
# 뒤에서 4개 중 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-4::2])
```

```python
# 뒤에서 앞으로 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::-2])
```

```python
# [5]위치부터 뒤로(-1) [1]위치 전까지
nums = [10, 20, 30, 40, 50, 60, 70]
print(nums[5:1:-1])
```

```python
# 마지막 5개를 역순으로
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-1:-6:-1])
```

```python
# 슬라이싱 할당(슬라이싱 구간에 대입문)
nums = [10, 20, 30, 40, 50]
nums[1:4] = [7, 8] # 1~3구간 [20, 30, 40]에 [7, 8]을 대입 
print(nums) # [10, 7, 8, 50]
```

```python
# 슬라이싱 결과를 다른 변수에 대입
a = [10, 20, 30, 40, 50, 60] 
b = a[:] # b = [10, 20, 30, 40, 50, 60]
c = a[::-1] # c = [60, 50, 40, 30, 20, 10]
d = a[1::2] # d = [20, 40, 60]
e = a[:2] # e = [10, 20]

print(b)
print(c)
print(d)
print(e)
```

```python
a = [10, 20, 30]
b = a

print(a)
print(b)
print(a == b)
print(a is b)

b[0] = 999
print(a) 
print(b) 
```

```python
a = [10, 20, 30]
b = a[:]

print(a)
print(b)
print(a == b) 
print(a is b)

b[0] = 999
print(a) # [10, 20, 30]
print(b) # [999, 20, 30]
```

```python
# 리스트 비교연산

a = [1, 2, 3]
b = [1, 2, 3]
c = [1, 2, 4]

print(a == b) # True
print(a == c) # False
```

```python
# 리스트를 인자값으로 자주 사용되는 내장함수

nums = [10, 20, 30, 40]

print(len(nums))
nums = [10, 20, 30, 40]

print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
# 리스트 메서드

# 리스트 객체는 자신만의 메서드를 가지며, 대부분의 리스트 메서드는 자기 자신을 수정하는데 사용된다.
# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
# extend() : 여러 개 추가
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
```

```python
# remove() : 값으로 삭제
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)

# 수정은 인덱스 할당, 슬라이싱 할당으로 수정
```

```python
# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)
```

```python
# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
# del 문
nums = [10, 20, 30, 40]
del nums[1]
print(nums)
```

```python
# reverse() : 순서 뒤집기, 정렬과 상관없이 현재 순서를 거꾸로 만든다.
nums = [10, 20, 30, 40]
nums.reverse()
print(nums)
```

```python
# count() : 특정 값이 몇개인지
nums = [10, 20, 10, 30, 10]
print(nums.count(10)) # 3
```

```python
# index() : 특정 값의 인덱스
fruits = ["apple", "banana", "orange"]
print(fruits.index("banana")) # 1
```

```python
# 리스트 정렬

# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
# 정렬의 경우 내장함수를 사용하여도 된다. sorted 내장함수는 리스트 뿐만 아니라 모든 컬렉션 타입을 정렬할 수 있고, 원본은 변경없이 정렬된 객체를 반환한다.

nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)

nums = [50, 10, 40, 20]
result = sorted(nums, reverse=True)
print(result)

fruits = ["banana", "apple", "orange"]
result = sorted(fruits)
print(result)

# key매개변수를 이용하여 문자열 길이로 정렬
fruits = ["watermelon", "kiwi", "banana", "apple"]
result = sorted(fruits, key=len)
print(result)

# 매개변수 key값으로는 함수가 온다. 뒤에서 람다함수를 배우면
# key값으로 람다식을 대입하여 다양한 요구 사항의 정렬도 가능하다.
# 자기 자신을 정렬할때는 sort()메서드, 다양한 형태의 정렬된 리스트를 반환할때는 sorted() 내장함수 사용
```
