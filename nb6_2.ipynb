# 컬렉션 타입의 반복
- 컬렉션객체는 for문을 사용하면 자신이 가진 아이템을 순환할 수 있다.

- 문은 값을 반환하지 않는다.

- 콜론문자(:)와 들여쓰기를 구문에 사용하여 함께 실행 되는 문들의 묶음인 코드 블록을 표현한다.

- 문은 표현(식)을 포함할 수 있다. for문은 in연산 표현식이 사용된다.

```python
# 리스트
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

```python
# 튜플
colors = ("red", "green", "blue")

for color in colors:
    print(color)
```

```python
# 셋
nums = {10, 20, 30}

for num in nums:
    print(num)
```

```python
# 딕셔너리
student = {
    "name" : "홍길동",
    "age" : 20
}

# 키
for key in student:
    print(key)

# 값
for value in student.values():
    print(value)

# 키 + 값 -> 언패킹
for key, value in student.items():
    print(key, value)


for key, *value in student.items():
    print(key, value)
```
