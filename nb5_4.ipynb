# 튜플
- 리스트와 비슷하지만, 소괄호()를 사용한다.(요소1, 요소2, ...., 요소n), 리스트의 경우 항목을 변경 가능하지만 튜플을 항목을 변경할 수 없다.

```python
# 튜플의 패킹(Packing), 언패킹(Unpacking)
t = 1, 3, 5 # 패킹

print(t) 
print(type(t))

l = list(t)
print(l, type(l))

x, y, z= t # 언팩킹
print(x, y, z)
print(type(x))
```

```python
# 리스트와 튜플간의 자료형 변경
x = (1, 3)
print(x)

y = list(x)
print(y)

z = tuple(y)
print(z)
```

```python
# 튜플 연산
a = (1, 2)
b = (3, 4)

print(a + b)
print((1, 2) * 3)
```

```python
t = (1, 2, 3)
print(2 in t)
print(5 in t)
```

```python
t = (10 ,20 ,30 ,40)
print(t[1])
print(t[1:3])
```

```python
# 튜플 메서드 관련 내장함수(튜플 ->리스트 변경 ->튜플 때문에 메서드가 많지 않음)
t = (1, 2 ,3, 4)

print(t.count(2))
print(t.index(2))
```

```python
t= (1, 2, 3)

lst = list(t)
lst.append(4)

t2 = tuple(lst)
print(t2)
```
