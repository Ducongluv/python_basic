```python
# enumerate()
names = ["홍길동", "이순신", "강감찬"]

for i, name in enumerate(names): # (0, "홍길동")
    print(f"({i + 1}) {name}")
```

```python
# reversed()
nums = [1, 2, 3]
print(list(reversed(nums)))
```

```python
# any()
nums = [0, 0, 3]
print(any(nums))

nums = ["", "2", "3"]
print(any(nums))

nums = [{}, {}, {}]
print(any(nums))
```

```python
# zip()

# zip()은 여러 개의 시퀀스 객체를 같은 위치(인덱스)의 값끼리 묶어주는 함수이다.

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
gender = ["F", "M", "F"]
result = zip(names, scores, gender)
print(list(result)) # [('홍길동', 90), ('이순신', 80), ('김유신', 100)]
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = {name : score for name, score in zip(names, scores)}

print(result)
```
