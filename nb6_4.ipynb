```python
# 조건표현식(삼항 연산자)

age = 20
result = "성인" if age >= 19 else "미성년자"

# bool ? 참값 : 거짓값
# a + b
# a++
print(result)
```

```python
# 조건문(if 문)
# 문은 값을 반환하지 않는다.

# 콜론문자(:)와 들여쓰기를 구문을 사용하여 함께 실행 되는 문들의 묶음인 코드 블록을 표현한다.
age = 20
if age >= 19:
    print("성인입니다.")
```

```python
score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")

```

```python
print(None, bool(None))
print(0, bool(0))
print(1, bool(1))
print([], bool([]))
print("",bool(""))
print("a", bool("a"))
```

```python
# bool은 사실 따로 없음
if bool(0):
    print("false")

if bool(1):
    print("test")

# 공백이 아니니 true 반환
if [1, 2]:
    print("Test")
```

```python
# 중첩
age = 20
score = 85

if age >= 19:
    if score >= 80:
        print("합격")
```

```python
num = 7

if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
num = -5

if num > 0:
    print("양수")
elif num < 0:
    print("음수")
else:
    print("0")
```

```python
scores = [95, 60, 80, 45]

for score in scores:
    if score >= 70:
        print("합격")
    else:
        print("불합격")
```

```python
numbers = [3, 8, 11, 20, 25]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} : 짝수")
    else:
        print(f"{num} : 홀수")
```

```python
fruits = ["apple", "banana", "kiwi", "orange"]

for fruit in fruits:
    if len(fruit) >= 6:
        print(fruit)
```
