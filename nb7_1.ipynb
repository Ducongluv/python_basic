```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

# pass keyword : python for문은 :로 시작 하지만 닫는게 없음 그래서 pass로 아무것도 없는 for문을 그냥 넘기는 것
# for i in list_nums:
#     pass

even = []
for i in list_nums:
    if i % 2 == 0:
        even.append(i)

print(even)

# 컴프리핸션
even2 = [i for i in list_nums if i % 2 == 0]

print(even2)

mul_10 = [i * 10 for i in list_nums if i <= 5]

print(mul_10)

name = ["홍길동", "이순신", "이황"]

print([n + "님" for n in name if n.startswith("이")])

name2 = ["홍길동", "이순신", "김유신", "박혁거세"]

print({n[:1]:n[1:] for n in name2})

id_pwd = ["admin-1234", "guest-0000", "zeroday-7777"]

print({ip.split("-")[0]:ip.split("-")[1] for ip in id_pwd})
print({i : p for i, p in (ip.split("-") for ip in id_pwd)})
```
