# SET : 중복을 허용하지 않는 자료형, 순서가 없음 (인덱싱 불가)

```python
a = {1, 2, 3}
b = {3, 4, 5}

# 수학용 집합 연산
print(a & b) # 교집합
print(a | b) # 합집합
print(a - b) # 차집합
```

```python
# 셋 내장함수
s = {1, 2, 3}

s.add(4) # 값 하나 추가 
print(s)

s.update([5, 6]) # 여러 개 추가

s.remove(2) # 값 2 삭제(없으면 오류)
print(s)

s.discard(10) # 값 없으면 그냥 무시
```

```python
s = {1, 2, 3}

print(len(s)) # 길이
print(2 in s) # 2 여부

s.clear() # 전체 삭제
print(s)
```

```python
fs = frozenset([1, 2, 3])
print(fs)
# fs.add(4) 오류남 읽기 전용 set은 수정 불가 
```

- {}은 빈 set이 아닌 빈 dict 자료형임
- 빈 셋이 필요하면 set() 함수 사용할 것
