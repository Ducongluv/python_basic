## 데이터 전처리

```python
# 결측치 확인
missing_count = 0

for student in students:
    if student["성적"] is None:
        missing_count += 1

print("결측치 개수 : ", missing_count)
```

```python
# 결측치 처리(평균값 대체 -> 할 때마다 요구사항에 따라 달라질 수 있음)
scores = []

for student in students:
    if student["성적"] is not None:
        scores.append(student["성적"])

avg = sum(scores) / len(scores)

for student in students:
    if student["성적"] is None:
        student["성적"] = avg

students
        
```

```python
# 이상치 탐지(정상 범위 : 0 ~ 100)
for student in students:
    score = student["성적"]

    if score < 0 or score > 100:
        print(student)
```

```python
# 이상치 처리
for student in students:

    if student["성적"] > 100:
        student["성적"] = 100
    
    elif student["성적"] < 0:
        student["성적"] = 0

students
```

```python
# 특성 추출(성별 구하기)
for student in students:

    gender_code = student["주민번호"].split("-")[1][0]

    if gender_code in ["1", "3"]:
        student["성별"] = "남"
    else:
        student["성별"] = "여"

students
```

```python
# 특성 추출(나이 구하기)
from datetime import datetime

current_year = datetime.now().year

for student in students:
    rrn = student["주민번호"]
    yy = int(rrn[:2])
    code = rrn.split("-")[1][0]

    if code in ["1", "2"]:
        birth_year = 1900 + yy
    else:
        birth_year = 2000 + yy

    student["나이"] = current_year - birth_year

students
```

## EDA : 탐색적 데이터 분석

```python
# 기술 통계
scores = []

for student in students:
    scores.append(student["성적"])

print("개수:", len(scores))
print("합계:", sum(scores))
print("평균:", sum(scores) / len(scores))
print("최대:", max(scores))
print("최소:", min(scores))
```

```python
# 성별 평균 성적
male_sum = 0
male_count = 0

female_sum = 0
female_count = 0

for student in students:

    if student["성별"] == "남":
        male_sum += student["성적"]
        male_count += 1

    else:
        female_sum += student["성적"]
        female_count += 1

print("남 평균:", male_sum / male_count)
print("여 평균:", female_sum / female_count)
```

```python
# 연령대별 평균 성적
age_group = {}

for student in students:

    decade = (student["나이"] // 10) * 10

    if decade not in age_group:
        age_group[decade] = []

    age_group[decade].append(student["성적"])

age_group
```

```python
for decade, scores in age_group.items():
    avg = sum(scores) / len(scores)
    print(f"{decade}대 평균 : {avg:.2f}")
```
