```python
a = 5
b = 2

print(a + b)   # 7
print(a // b)  # 2
print(a % b)   # 1
print(5 ** 2)
```

```python
print(5 > 3)      # True
print(5 == 3)     # False
```

```python
print(True and False)   # False
print(True or False)    # True
print(not True)         # False

print(10 > 5 and 10 == 7)
```

```python
print(3 in [1, 2, 3])       # True
a = [1, 2, 3]
print(4 in a)  # False
print(4 not in a)   # True
```

```python
a = [1, 2]
b = a

a is b      # True
a == b      # True

c = [1, 2]
d = [1, 2]

c is d      # False
c == d      # True

# == : 객체의 값만 비교 

# is : 객체 비교
```

```python
score = 50
result = "합격" if score >= 60 else "불합격"
print(result)
```
