## with 문
: 파일이나 자원(resource : 프로그램이 시스템에게서 빌려서 사용하는 것) 등을 사용한 위 자동으로 정리(닫기)해 주는 문법

```python
# with 자원해지가 필요한 객체 as 변수:
#     실행할 코드
```

```python
# with문 사용하지 않는 경우
f = open("students.csv", "r", encoding="utf-8")
print(f.read())
f.close()   # 직접 닫아야 함. 실수로 빼먹거나, 실행전 에러가 발생하면 메모리 누수가 발생한다. 

# with문을 사용하는 경우
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
```

```python
with open("students.txt", "r", encoding = "utf-8") as f:
    # 한 줄씩 읽기
    for line in f:
        print(line) # 라인 끝의 줄바꿈이 문자 포함되어 두번 줄바꿈이 일어남
```

```python
with open("students.txt", "r", encoding = "utf-8") as f:
    # 한 줄씩 읽기
    for line in f:
        print(line.strip())
```

```python
a = "홍길동,90".split(",")
a[0] = "홍길동"
a[1] = 90

a, b = "홍길동,90".split(",")
print(a)
print(b)
```

```python
names = []
scores = []
with open("students.txt", "r", encoding = "utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")

        # print(name, int(score))
        names.append(name)
        scores.append(score)

print(names)
print(scores)
```

```python
# 딕셔너리 만들기
students = {}

with open("students.txt", "r", encoding = "utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```

## 2.csv 파일

```python
import csv

# open()은 내장함수
with open("students.csv", "r", encoding = "utf-8") as f:
    # reader() : csv모듈안에 있는 함수
    reader = csv.reader(f)

    for row in reader:
        print(row)
        print(type(row))
```

```python
with open("students.csv", "r", encoding = "utf-8") as f:
    reader = csv.reader(f)

    next(reader)

    for row in reader:
        print(row)
```

```python
# DictReader()를 사용하여 첫번째 줄을 키로 하는 딕셔너리로 반환
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인 키, 나머지 내용 값으로

    for row in reader:
        print(row)
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) 

    data = list(reader)

print(data)

data[2]["영어"] = 80

print(data[2])
```

## ​3. CSV 파일로 저장

csv.DictReader()로 읽어서 data를 수정했다면, csv.DictWriter() 또는 일반 파일 쓰기를 이용해 다시 저장하면 된다. 

```python
import csv

# students_result.csv 파일을 쓰기(w) 모드로 열기
# newline="" : 빈 줄이 추가되는 현상 방지
# encoding="utf-8-sig" : 한글이 엑셀에서 깨지지 않도록 저장
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
