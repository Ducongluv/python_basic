## 람다함수 문법

​

함수 정의는 함수선언문 이다 - 함수를 정의하는 문

람다는 함수 표현식 이다 - 람다 표현식의 결과로 함수를 반환한다.

​

※ 파이썬에는

조건문 / 조건표현식(삼항연산자) 

반복문 / 반복표현식(컴프리핸션)

함수(문) / 람다(함수표현식)

가 있다.



```python
def plus(x):
    return x + 10

a = lambda x : x + 10 # 변수 저장 -> 1급함수기능포함

a(50)
```

```python
# 일반 함수 정의
def add1(a, b):
    return a + b

# 람다 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)

r2 = add2(1, 3)
print(r2)
```

```python
def make(n, func): # n:숫자, func:함수-
    return func(n)
```

```python
def f1(n):
    return n*n

def f2(n):
    return n*2

def f3(n):
    return "짝수" if n%2==0 else "홀수" 
```

```python
r = make(10, f3)
print(r)
```

## 람다함수 즉시 호출 

​

람다의 결과값이 함수이다. 람다의 결과값에 () 호출연산자를 사용하면 람다가 호출된다

```python
print((lambda x, y : x * y)(5, 10))
```

```python
s = [('홍길동', 90), ('이순신', 85), ('강감찬', 100)]
r = sorted(s, key=lambda x: x[1], reverse=True)
# lambda x: x[1] 
# x(리스트의 요소 튜플)를 입력받아 x[1](튜플의 인덱스1값)를 반환한다.
print(r)
# [('강감찬', 100), ('홍길동', 90), ('이순신', 85)] 점수의 내림차순으로 정렬됨

# 원본 리스트를 하나씩 꺼낸다.
('홍길동', 90)
('이순신', 85)
('강감찬', 100)

# ② key 함수의 인수로 하나씩(리스트 사이즈 만큼 key함수를 반복 호출하면서 하나씩 전달)  전달한다.

# ③ sorted 내장함수는 인수로 입력받은 key함수의  반환값(x[1] 요소)으로 정렬한다.
```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)

# r2 = sorted(s, key = score, reverse = False)
r2 = sorted(s, key = lambda e:e[1], reverse = False)
print(r2)
```

## 람다를 인자로 받는 내장 함수 중 자주 사용되는 것들

함수(주로 lambda)를 인수로 받는 함수를 고차 함수(Higher-order Function)라 한다.

```python
# map() - 모든 요소를 변환
nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = list(map(lambda x : x * 2, nums))

print(result)
```

```python
# filter() - 조건에 맞는 요소만 선택
nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = list(filter(lambda x : x % 2 == 0, nums))

print(result)
```

```python
# sorted() - 정렬 기준 지정
names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key = lambda x : len(x), reverse = True)

print(result)
```

```python
students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["나이"])
print(result)

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
# "점수"를 기준으로 내림차순 정렬
result = sorted(students, key=lambda x: x["점수"], reverse=True)
print(result)
```

```python
# "이름"을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["이름"])
print(result)
```

```python
# 여러 키를 기준으로 정렬
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]), reverse = False)
print(result)
```

```python
# min() - 기준에 따라 최소값 찾기

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

# 알파벳 순
# result = min(names, key=lambda x: x)
print(result) # kiwi
```

```python
# max() - 기준에 따라 최대값 찾기

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 긴 요소 반환
result = max(names, key=lambda x: len(x))

print(result)
```

```python
# any() - 하나라도 조건을 만족하는지 확인

nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환
result = any(map(lambda x: x % 2 == 0, nums))

print(result)
```

```python
nums = [1, 3, 5, 8]
# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환

r1 = map(lambda x : x % 2 == 0, nums)
print("r1 : ", r1)
r2 = list(r1)
print("r2 : ",r2)

result = any(map(lambda x : x % 2 == 0, nums))

print(result)
```

```python
# all() - 모두 조건을 만족하는지 확인
nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result)
```
