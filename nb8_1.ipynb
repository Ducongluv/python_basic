## 1. while 반복문
- while문은 헤드라인에 있는 테스트가 참 값으로 평가되는 한 문의 블록을 반복해서 실행
- while문은 테스트가 거짓이 될 때까지 문의 시작으로 돌아오도록 제어가 유지되어 Loop라 부름
- 테스트가 거짓이 되면, 제어는 while 블록 다음 문으로 넘어감

```python
greet = "Hello"
while greet: # 문자열의 경우 공백이면 False, 아니면 True
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10
# 다중 할당
# 내부적으로 오른쪽의 튜플 (0, 10)울 왼쪽의 변수들에 언패킹해서 각각 저장

while a < b:
    print(a)
    a = a + 1
```

## 2. for 반복문
- 정렬된 시퀀스 또는 가변 객체 안의 아이템들을 순차적으로 처리
- 다음과 같이 단계별 처리하고자 하는 객체와 함께 할당 대상을 헤드라인에 명시, 반복하고자 하는 문들의 블록이 따라옴

```python
# for 할당변수 in 객체:
#     statements
# else:
#     statements
for a in [5, 10, 15]:
    print(a)
```

```python
for (a, b) in [(1, 2), (10, 20)]:
    print(a)
    print(b)
```

```python
for a, *b, c in [(1, 2, 3, 4), (5, 6, 7, 8)]:
    print(a, b, c)
```

## 3. break, continue, pass, loop else
- break : 반복문을 즉시 종류
- continue :  현재 반복만 건너뛰고 다음 반복 수행
- pass : 아무 동작도 하지 않음
- else : break 없이 반복문이 정상 종료되면 실행

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)

print("반복 종료")
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i) 
```

```python
# 루프 else
# break가 없는 경우
for i in range(3):
    print(i)
else:
    print("정상 종료")
```
